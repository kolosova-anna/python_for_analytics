### 1.1. 
Создать новый признак price_per_sq_lot, который будет содержатьстоимость за один квадратный метр общей площади

```python
df['price_per_sq_lot'] = (df['price']/df['sqft_lot']).round(2)
df.sample(5)
```

### 1.2
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

```python
def delta_renov(x, y):
    if y != 0:
        return y - x
    else:
        return 0
df['delta_renovated'] = df.apply(lambda x: delta_renov(x['yr_built'], x['yr_renovated']), axis =  1)
df.sample(10)
```

### 1.3
Создайте признаки года продажи, месяца продажи

```python
df['yr_sale'] = pd.DatetimeIndex(df['date']).year
df['mth_sale'] = pd.DatetimeIndex(df['date']).month
df.sample(10)
```

### 1.4
Удалите признаки date, zipcode, lat, long

```python
df.drop(columns = ['date', 'zipcode', 'lat', 'long'], inplace = True)
df.head()
```

## 3 задача
Составьте несколько сводных таблиц
### 3.1
Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
df.pivot_table(index = 'bedrooms',
              values = 'price',
              aggfunc = 'mean').sort_values('price')
```

### 3.2
Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
grouped = df.groupby('condition')
grouped.agg({'price': ['min', 'mean', 'max']})
```

### 3.3
Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.pivot_table(index = 'waterfront',
              columns = 'view',
              values = 'price',
               aggfunc = 'count',
               margins = True,
              margins_name = 'Total')
```

### 3.4
Каких домов в зависимости от этажности и количества спален больше?

```python
df.pivot_table(index = 'floors',
              columns = 'bedrooms',
               values = 'price',
               aggfunc = 'count')
```

_Ответ:_ Больше всего одноэтажных домов с тремя спальнями.

### 3.5
Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
pd.crosstab(index = df['condition'],
           columns = df['grade'],
           values = df['price'],
           aggfunc = 'median')
```
